知能システム理論基礎の授業及び演習と自分の理解を通じて、私はナイーブベイズモデルは他の分類手法により誤差率が低いと思うが、ナイーブベイズの原理を考えると、ナイーブベイズモデルが各特徴がお互いに独立していることを前提と仮定している。ベイズの定理を用いて、各独立事相の確率により分類の確率を計算する。つまり、各特徴量が独立であるとき有効である。しかし、実際の分類問題中のデータには、各特徴が相関関係を持つが多いと思う、もし各特徴量お互いに関連性が高い特徴量が多い場合、ナイーブベイズモデルの有効性が低下となると考えている。

**課題の核心部分のコード：**

```python
from janome.tokenizer import Tokenizer
import numpy as np
```

```python
class HAONaiveBayes():
    #テキスト入力部
    def __init__(self, file1, file2):
        f = open(file1)
        self.text1 = f.read()
        f = open(file2)
        self.text2 = f.read()

        full_text = self.text1 + self.text1

        self.text1Dic = self.split_words(self.text1)
        self.text2Dic = self.split_words(self.text2)
    #テキスト分割部
    def split_words(self,text):
        t = Tokenizer()
        tokens = t.tokenize(text)
        List =  [token.surface for token in tokens]
        
        text = {}
        for i in List:
            if i not in text.keys():
                text[i] = 1
            else:
                text[i] += 1
        return text
    #ナイーブベイズ予測部
    def NaiveBayes(self,textTest):
        dic1 = self.text1Dic
        dic2 = self.text2Dic
        #辞書分割により初期化
        dicTest = self.split_words(textTest)
        #テキスト１とテキスト２に所属する確率の初期化
        P1 = 1/2
        P2 = 1/2

        for j in dicTest.keys():
            if j not in dic1.keys() or j not in dic2.keys():
                continue
        #ナイーブベイズにより確率を計算する
            P1 = P1 * np.log(dic1[j] / sum(dic1.values()))
            P2 = P2 * np.log(dic2[j] / sum(dic2.values()))

        #予測の出力
        P = P1/(P1+P2),P2/(P1+P2)
        return P
    
    def print_file1(self): #テキスト１の出力
        print(self.text1)
    def print_file2(self): #テキスト２の出力
        print(self.text2)
```

```python
nb = HAONaiveBayes("ngsk.txt", "wiki.txt")
```

```python
nb.print_file1()
```

```python
nb.print_file2()
```

```python
test1 = "第二葉の写真の顔は、これはまた、びっくりするくらいひどく変貌していた。"
print(nb.NaiveBayes(test1))
```

```python
test2 = "抽象的には、分類器の確率モデルは次のような依存クラス変数Cについての条件付きモデルである。"
print(nb.NaiveBayes(test2))
```

```python
test3 = "斜め七十七度の並びで泣く泣くいななくナナハン七台難なく並べて長眺め"
print(nb.NaiveBayes(test3))
```

　テストテキスト2は約52.5%の確率で訓練データテキスト2の「人間失格」に所属する、47.5%確率で訓練データテキスト2の「単純ベイズ分類器の解説」に所属すると予測したが、実際には訓練データテキスト2の「単純ベイズ分類器の解説」に所属する。

　両方とも所属していないテストテキスト3は本モデルにより、約48.1%の確率で訓練データテキスト1の「人間失格」に所属する、約51.9%の確率で訓練データテキスト3の「単純ベイズ分類器の解説」に所属する。
